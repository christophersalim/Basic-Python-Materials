## Programming in Python

In Python, there are several programming logic that can be used, just like other programming languages. The first example is if logic. If logic is used when we want to apply conditions to our program.

```python
# simple example with one condition
x = 2 # initiate x value
if x > 1: # if x > 1, execute this line below. Otherwise, skip.
    print("change the number!")
```

If you pay attention, different things from MATLAB are:

1. Python uses:, whereas MATLAB does not.

2. In Python it doesn't need to end with end in every loop.

3. In MATLAB, indenting is optional. However, in Python, it is necessary to indent 1 tab or 4 spaces each to loop. Indenting is useful for clearly knowing the location of a program in a loop, because often a program can have many loops.

```python
# example with two conditions
x = 3
if x >= 3: # execute this line below is x is larger or equal to 3
    print('just right.')
else: # execute this line below if otherwise.
    print('no')
```

```python
# example with three conditions
x = 1
if x >=3:
    print('greater than 3')
elif 2<x<3:
    print('between two and three')
else:
    print('I dont understand')
```

```python
# a slightly more complicated example

demam = True
flu = False

if demam == True: # execute this line if demam is True
    if flu == True: # if demam and flu are True
        print('perlu dibawa ke dokter!')
    else: # if demam True, flu False
        print('mungkin perlu')
else: # execute this if demam is False
    if flu == False:
        print('sehat')
    else:
        print('tidak tahu')
```

```python
x = 'demam'
y = 10 if x == 'demam' else 0
print(y)
```

Next is programming using a while loop. While this also functions as a conditional. However, unlike the if loop which provides conditions for the program to run, while the loop using conditions for the program can continue as long as those conditions are met.

```python
# simple example with one while only

topher = 1
while topher < 10:
    print('this is '+ str(topher))
    topher += 1
```

However, using a while loop can be dangerous if not done with good understanding. Why do you think the script below can't stop?

```python
chris = 1

while chris < 10:
    print(chris)
```

```python
# second example with while loop 
x = -20
y = 20
while x <= y:
    print('Nilai x: ', x)
    x = x + 1
```

```python
# nested while loop
x = -20
y = 20
while x <= y:
    print('X is now: ', x)
    x = x + 1
    while x <= 0:
        print('X is negative')
        x = x + 1
        while x <= -1:
            print('x kurang dari -1')
            x += 1
```

From the examples above, you can already see how the while loop works. Now, try to do the while loop programming combined with the for loop.

```python
for k in range(0,9):
    x = k + 1
print(x)
```

```python
x = []
for k in range(0,9):
    a = k + 1
    x.append(a)
print(x)
```
